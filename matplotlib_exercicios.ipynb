* O comando linspace é uma função da biblioteca NumPy, que cria uma sequência de números igualmente espaçados em um intervalo especificado. A função linspace recebe três argumentos: o valor inicial, o valor final e o número de elementos que você deseja na sequência.

```python
# Import o NumPy
import numpy as np
# Crie um array de 100 números (distribuídos normalmente) entre 0 e 100
# salvando o array em uma variável x
x = np.linspace(0, 1, 100)
```

```python
# Crie um plot usando plt.subplots() e plot x e x² (squared)
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

* Scatter = gráfico de dispersão

```python
# Crie uma scatter plot de X e X exponencial (np.exp(X))
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Crie um dicionário Python de 3 produtos
# as chaves do dicionário podem ser nomes de alimentos
# e os valores podem ser preços
produtos = {"Arroz": 5.99,
            "Feijão": 10.99,
            "Carne": 20.99}
```

```python
# Crie um gráfio de barras onde x-axis recebe como valor as chaves do dicionário
# e y-axis recebe os valores do dicionário
fig, ax = plt.subplots()

ax.bar(produtos.keys(), produtos.values())
# Adicione um título, xlabel e ylabel ao último plot criado
ax.set(title = "Produtos", ylabel = "Preços");
```

```python
# Faça o mesmo plot, dessa vez com barras horizontais
fig, ax = plt.subplots()
ax.barh(list(produtos.keys()),list(produtos.values()));
```
